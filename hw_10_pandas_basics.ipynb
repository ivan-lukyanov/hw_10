#### Задание 1
Скачайте с сайта grouplens.org...movielens/ датасет любого размера.   
Определите какому фильму было выставлено больше всего оценок 5.0.

```python
# портирую библиотеку pandas и присваиваиваю ей псевдоним pd

import pandas as pd
```

```python
# читаю данные из csv

data_rating = pd.read_csv('10/ml-latest-small/ratings.csv')
data_rating.head()
```

```python
data_movies = pd.read_csv('10/ml-latest-small/movies.csv')
data_movies.head()
```

```python
# объединяю файлы csv (в одном есть название фильма, в другом рейтинг), 
# объединяю по столбцу 'movieId'^ за счет параметров left_on и right_on

data_rating_movies = data_rating.merge(data_movies, left_on='movieId', right_on='movieId')
data_rating_movies.head()
```

```python
# вычисляю оценку 5 (посчитал что объективнее взять расчетный результат и полжитьего в переменную)

rating_five = data_rating['rating'].max()
print(rating_five)
```

```python
# отфильтровываю данные что бы получить все фильмы с рейтингом 5

movies_rating_five = data_rating_movies[data_rating_movies['rating'] == rating_five]
movies_rating_five.head()
```

```python
# подсчитываю уникальные значения заголовков фильмов с оценкой 5. 
# по умолчанию максимальное значение на верху 
# ограничиваю искомое значение еhead() и получаю ответ

movies_max_rating_five = movies_rating_five['title'].value_counts().head(1)
movies_max_rating_five
```

### Задание 2
По данным файла power.csv  
- Посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 

- За период с 2005 по 2010 года.  

- Не учитывайте в расчетах отрицательные значения quantity.

```python
# создаю переменную с дата фреймом за счет метода read_csv  
# Для просмотра формата страници использую head()

df_power = pd.read_csv('10/power.csv')
df_power.head(3)
```

```python
# Вариант_1

# Создаю новый дата фрейм, но с отфильтрованными строками методом isin
# Фильтры ставлю на year и на category
# в конце считаю сумму данных столбца quantity, которые соответсвуют отфильтрованным строкам в столбцах year и category

sum_quantity_power = df_power[df_power['year'].isin([2005, 2006, 2007, 2008, 2009, 2010])& 
                              df_power['category'].isin([4, 12, 21])] ['quantity'].sum()
sum_quantity_power
```

```python
# Вариант_2

# отличемтся от 1го тем, что года в year не перечисляются, а указывается фильтр на вхождение в диапазон a формате от и до

sum_quantity_power = df_power[df_power['year'].between(2005, 2010, inclusive=True)& 
                              df_power['category'].isin([4, 12, 21])] ['quantity'].sum()
sum_quantity_power
```
